```python
import numpy as np
import pandas as pd

input_file = "kc_house_data.csv"
df = pd.read_csv(input_file)
df.info()
```

```python
housing_prepared = df.drop(["id", "price", "date", "zipcode"], axis=1)  # id, date, zipcode, price 특성 제거.
housing_labels = df["price"].copy()
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_scores = cross_val_score(lin_reg, housing_prepared, housing_labels, scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)
display_scores(lin_rmse_scores)
```

## Feature Engineering Try!


```python
from sklearn.preprocessing import FunctionTransformer

living_ix, living15_ix, lot_ix, lot15_ix = [
    list(df.columns).index(col)
    for col in ("sqft_living", "sqft_living15", "sqft_lot", "sqft_lot15")]

def add_extra_features(X):
    sqft_living_ratio = X[:, living_ix] / X[:, living15_ix]
    sqft_lot_ratio = X[:, lot_ix] / X[:, lot15_ix]
    return np.c_[X, sqft_living_ratio, sqft_lot_ratio]

attr_adder = FunctionTransformer(add_extra_features, validate=False)
housing_extra_attribs = attr_adder.fit_transform(df.values)
```

```python
import pandas as pd

housing_extra_attribs = pd.DataFrame(
    housing_extra_attribs,
    columns=list(df.columns)+["sqft_living_ratio", "sqft_lot_ratio"],
    index=df.index)
housing_extra_attribs
```

```python
housing_prepared_new = housing_extra_attribs.drop(["price", "date", "zipcode", "id"], axis=1)  # 특성 추출을 한 housing_prepared_new
housing_prepared_new
```

## Using GradientBoostingRegressor (Min)

```python
from sklearn.ensemble import GradientBoostingRegressor
gdb_reg = GradientBoostingRegressor(n_estimators=3000, learning_rate=0.05, max_depth=3, max_features='sqrt',
                                          min_samples_leaf=15, min_samples_split=10, loss='huber')
gdb_scores = cross_val_score(gdb_reg, housing_prepared, housing_labels, scoring="neg_mean_squared_error", cv=10)
gdb_rmse_scores = np.sqrt(-gdb_scores)
display_scores(gdb_rmse_scores)
```

# GridSearchCV를 이용한 Gradient Boosting 하이퍼 파라미터 튜닝

```python
from sklearn.model_selection import GridSearchCV

params = {
    'learning_rate': [0.05, 0.1, 0.5, 1],
    'max_depth' : [3, 4, 5],
    'n_estimators': [3000, 5000],
    'min_samples_leaf': [10, 15, 20]
}

gb_reg = GradientBoostingRegressor(random_state=42)
gb_reg.fit(housing_prepared, housing_labels)
```

```python
grid_cv = GridSearchCV(gb_reg, param_grid = params, cv=3, verbose = 1)
grid_cv.fit(housing_prepared, housing_labels)
print('최적 하이퍼 파라미터:', grid_cv.best_params)
print('최고 예측 정확도: {0:.4f}'.format(grid_cv.best_score_))
```

# 위에 변수명 best_params_가 아니라 오류 뜰거임. 위 실행 말고 밑에 코드 실행
